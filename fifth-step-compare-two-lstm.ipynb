```python
file = open("exports/global.csv", "r", encoding = "utf8")
g = file.readlines()

file = open("exports/john.csv", "r", encoding = "utf8")
j = file.readlines()

file = open("data/pb-cleaned.txt", "r", encoding = "utf8")
txt = file.readlines()
words = txt[0].split(' ')
```

```python
def compareWords(index,j,g):
    yy = 0
    if(j[index] != g[index]):
        a = j[index].split(",")
        b = g[index].split(",")
        c = 0
        for x in range(len(a)):
            if(a[x] not in b):
                c = c + 1
        if c > 9:
            yy = yy + 1
            #print(str(i) + ": " + j[i])
        return c, j[index].split(",")[1]
    else:
        return 0
    
        
        
```

```python
def findWordIndex(word):
    word = word.lower()
    
    for i in range(27783):
        if j[i].split(",")[1] == word:
            return i
        
```

```python
def findWordIndexes(word):
    lines = []
    word = word.lower()
    
    for i in range(27783):
        for t in j[i].split(","):
            if t == word:
                lines.append(i)

    return lines
        
```

```python
def findWordIndexesG(word):
    lines = []
    word = word.lower()
    
    for i in range(27783):
        for t in g[i].split(","):
            if t == word:
                lines.append(i)

    return lines
        
```
